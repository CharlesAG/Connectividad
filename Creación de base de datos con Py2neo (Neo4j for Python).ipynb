Se requiere instalar la versión 4.2.2 de py2neo

```python
from py2neo import Graph
import json
from py2neo import Node, Relationship
```

```python
#Cambiar la contraseña a la correspondiente con el ordenador, en mi caso es "progonmath".
graph = Graph("bolt://localhost:7687", user="neo4j", password="progonmath")
```

```python
with open("contensiosos.json","r") as c:
    cont=json.load(c)
```

```python
L=[]
for i in range(0,len(cont["nodes"])):
    L.append(cont["nodes"][i]["id"])
L.sort()   
```

### Creación de nodos 

Nodos con información recabada:


```python
x=graph.begin()
for i in range(0,374):
    n=Node("NODOS",country=cont["nodes"][i]["country"],case_id=cont["nodes"][i]["case_id"],year=cont["nodes"][i]["year"],
           name=cont["nodes"][i]["name"],id_=cont["nodes"][i]["id"],type_=cont["nodes"][i]["type"],
           title=cont["nodes"][i]["DB"]["title"],txt=cont["nodes"][i]["DB"]["txt"],
          doc_type=cont["nodes"][i]["DB"]["doc_type"],source_doc=cont["nodes"][i]["DB"]["source_doc"],
          doc=cont["nodes"][i]["DB"]["doc"],date_creation=cont["nodes"][i]["DB"]["date_creation"],
          pdf=cont["nodes"][i]["DB"]["pdf"],date_modification=cont["nodes"][i]["DB"]["date_modification"],
          source_pdf=cont["nodes"][i]["DB"]["source_pdf"],error=cont["nodes"][i]["DB"]["meta_name"]["error"],
          actions=cont["nodes"][i]["DB"]["meta_name"]["actions"], date_sentence=cont["nodes"][i]["DB"]["meta_name"]["date_sentence"],
          court=cont["nodes"][i]["DB"]["meta_name"]["court"],number=cont["nodes"][i]["DB"]["meta_name"]["number"],
          involved_a=cont["nodes"][i]["DB"]["meta_name"]["involved"]["involved_a"],
          involved_b=cont["nodes"][i]["DB"]["meta_name"]["involved"]["involved_b"])
    x.create(n)
x.commit()
    
```

Nodos restantes:

```python
x=graph.begin()
for i in range(374,len(cont["nodes"])):
    n=Node("NODOS",name=cont["nodes"][i]["name"],id_=cont["nodes"][i]["id"],type_=cont["nodes"][i]["type"])
    x.create(n)
x.commit()
```

###  Creación de aristas con peso

```python
x=graph.begin()
for i in range(0,len(cont["links"])):
    ori=cont["links"][i]["ori_val"]
    tar=cont["links"][i]["target"]
    val=cont["links"][i]["value"]
    if ori in L:
        ori_n=graph.run("MATCH (n:NODOS) WHERE n.id_=$ori return n",ori=ori).data()
        tar_n=graph.run("MATCH (s:NODOS) WHERE s.id_=$tar return s",tar=tar).data()
        a=ori_n[0]["n"]
        b=tar_n[0]["s"]
        R=Relationship(a,"Incide ",b,value=val)
        x.create(R)
        print(f"link {i} created from ori_val={ori} to target={tar}, with value={val}")
    else:
        pass
    
    
x.commit()
```
